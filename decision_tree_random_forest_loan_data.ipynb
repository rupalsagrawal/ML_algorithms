now for applying a model we wnat to remove the categoricall columns so we use get_dummies and do this.

```python
final_data=pd.get_dummies(df,columns=['purpose'],drop_first=True)
final_data.head()
```

```python
from sklearn.model_selection import train_test_split
x=final_data.drop('not.fully.paid',axis=1)
y=df['not.fully.paid']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(x_train,y_train)
pred=tree.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=600)
forest.fit(x_train,y_train)
pred_forest=forest.predict(x_test)
print(confusion_matrix(y_test,pred_forest))
print(classification_report(y_test,pred_forest))
```
